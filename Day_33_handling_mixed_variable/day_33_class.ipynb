```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv") # sir has created this dataset
```

```python
df.drop(columns ='Ticket' , inplace =True)
```

```python
df.sample(5)
```

```python
df['number'].unique()
```

```python
fig = df["number"].value_counts().plot.bar()
fig.set_title("Passanger travelling with __ menbers")
```

```python
# extracting the numerrical parts 
df['number_numerical'] = pd.to_numeric(df['number'] ,errors='coerce',downcast='integer')
```

```python
# extracting categorical parts
df['number_categorical'] = np.where(df['number_numerical'].isnull(),df['number'],np.nan)
```

```python
df.head()
```

```python
# we can also replace the NaN value of number_numerical with 0 as zerp person is travling with them
```

```python
df['number_numerical'] = df['number_numerical'].fillna(0)
```

```python
df.head()
```

```python
# now separating the value from cabin column
df['Cabin_num'] = df['Cabin'].str.extract(r'(\d+)') # capture the numerical parts 
df['Cabin_cate'] = df['Cabin'].str[0] # capturing the numberical parts 
```
