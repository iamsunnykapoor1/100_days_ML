```python
def rename(name) :
    if name.lower() == "royal challengers bangalore" :
        return "RCB"
    else :
        return name
        
rename("Royal Challengers Bangalore")    
```

```python
pd.read_csv("IPL Matches 2008-2020.csv",converters={"team1":rename,"team2" :rename}).head()
```
