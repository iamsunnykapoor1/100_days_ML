now doing this process in the loop so that we get a large dataset of the movie

```python
df = pd.DataFrame()
```

```python
for i in range (1,532) :
    if response.status_code != 200 : # agar koi page per data nahi hai tu errorn nahi dega
        continue
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=e47c141b5a5a5101df12ad52991e073f&language=en-US&page={i}')
    temp_df = pd.DataFrame(response.json()['results'])[['id','title','release_date','vote_count','popularity']]
    df = pd.concat([df, temp_df], ignore_index=True)
```
