<h3>Doing differnt types of transformation on diffrent column together </h3>







<b>1.gender column -> OneHotEncoding

<b>2.fever column -> Simple imputer

<b>3.cough column -> OrdinalEncoder

<b>4.city column ->OneHotEncoder

we will be  doing  all these together 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop(columns = ["has_covid"]),
                                                 df["has_covid"],
                                                 test_size =0.2)
```

<h1>Aam Zindagi</h1>

<h6>doing the above transformation without Column Transfotmation class

```python
# adding simple imputer to  fever col
si = SimpleImputer()
x_train_fever = si.fit_transform(x_train[['fever']])

#also the test data
x_test_fever = si.fit_transform(x_test[['fever']])
```

```python
x_train_fever.shape
```

```python
#.cough column -> OrdinalEncoder
oe = OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough = oe.fit_transform(x_train[['cough']])

# also on the tset data
x_test_cough = oe.fit_transform(x_test[['cough']])
```

```python
x_train_cough.shape
```

```python
# OneHotEncoding -> gender, city
ohe = OneHotEncoder(drop ='first', sparse_output = False)
x_train_gender_city = ohe.fit_transform(x_train[['gender','city']])

# also for the test data
x_test_gender_city = ohe.fit_transform(x_test[['gender','city']])
```

```python
x_train_gender_city.shape
```

```python
# extracting the age column 
x_train_age = x_train.drop(columns =['gender','fever','cough','city']).values


x_test_age = x_test.drop(columns =['gender','fever','cough','city']).values
```

```python
x_train_age.shape
```

Now we have all the columns , we have to concatinate these column

```python
x_train_transformed = np.concatenate((x_train_age,x_train_fever,
                                     x_train_gender_city,
                                     x_train_cough),axis =1)
# also the test data
x_test_transformed = np.concatenate((x_test_age,x_test_fever,
                                     x_test_gender_city,
                                     x_test_cough),axis =1)
```

```python
x_train_transformed.shape
```

<h1>Mentos Zindagi</h1>

<h6>Now we are doing the same things with columnTransformation class </h6>

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output =False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
transformer.fit_transform(x_train).shape
```
